```python
plt.figure(figsize=(15, 12))
sns.histplot(df['Distance'], kde=True, bins=10)
plt.title(f'histogram of {col}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 12))
sns.violinplot(x=df['Distance'])
plt.title(f'violin plot of {col}')
plt.tight_layout()
plt.show()
```
